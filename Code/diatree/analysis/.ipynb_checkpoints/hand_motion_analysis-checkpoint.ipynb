```python
import os
import mumodo.mumodoIO as mio
from mumodo.analysis import *
from mumodo.InstantIO import *
import numpy as np
import pandas as pd
%matplotlib inline
```

### load hand motion data

```python
start_time = {'pr2_12112015': 1447324869939, 'pr3_18112015': 1447848910139}
```

```python
this_run = 'pr3_18112015'
data_path = os.path.join('..','data', this_run, this_run+'.xio.gz')
```

```python
hand_motion = mio.open_streamframe_from_xiofile(data_path, sensorname='VeniceHubReplay/Leap/Leapdata')
hand_motion = hand_motion.dropna()
```

```python
hand_motion = hand_motion[hand_motion.HandNum > 0]
```

```python
hand_motion;
```

### synchronize hand motion data and video

```python
t_start = start_time[this_run]
hand_motion = hand_motion[hand_motion.time > t_start]
hand_motion.index = hand_motion['time'].map(lambda x: x - t_start)
```

```python
f = lambda x: x[1][0] if x[0][0] == key else x[1][1] if x[0][1] == key else np.nan
```

```python
key = 'Left hand'
hand_motion['leftPV'] = list(hand_motion[['HandType', 'PalmVelocity']].apply(f, axis=1))
key = 'Right hand'
hand_motion['rightPV'] = list(hand_motion[['HandType', 'PalmVelocity']].apply(f, axis=1))
```

```python
left = hand_motion.ix[:, ['leftPV']]
right = hand_motion.ix[:, ['rightPV']]
```

```python
left = left.dropna()
left['leftVM'] = left['leftPV'].map(lambda x: np.linalg.norm([x[0], x[1], x[2]]))
right = right.dropna()
right['rightVM'] = right['rightPV'].map(lambda x: np.linalg.norm([x[0], x[1], x[2]]))
```

```python
left['leftVM'].to_csv(os.path.join('..', 'data', this_run, 'left_velocity.csv'))
```

```python
right['rightVM'].to_csv(os.path.join('..', 'data', this_run, 'right_velocity.csv'))
```
